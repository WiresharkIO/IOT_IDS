```python
import matplotlib.pyplot as plt
```

```python
# List of model names
models = ['Naive_Bayes', 'Naive_Bayes_AdaBoost', 'Random_Forest', 'SVM', 'MLP']
accuracies = [0.39822587303771606, 0.8764521882724053, 0.8199938604198203, 0.696046662346079, 0.5338]
colors = ['red', 'green', 'blue', 'cyan'] 

plt.figure(figsize=(7, 6))
plt.bar(models, accuracies, color=colors)
plt.title('Accuracy Comparison of Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.ylim(0.0, 1.0)
plt.show()
```

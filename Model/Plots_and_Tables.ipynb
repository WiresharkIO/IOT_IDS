```python
import matplotlib.pyplot as plt
```

```python
# List of model names
models = ['Naive_Bayes', 'Random_Forest', 'SVM', 'MLP']
accuracies = [0.39822587303771606, 0.9894145603802117, 0.696046662346079, 0.9834]
colors = ['red', 'green', 'blue', 'cyan'] 

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=colors)
plt.title('Accuracy Comparison of Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0.0, 1.0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# List of model names
models = ['Naive_Bayes', 'Naive_Bayes_AdaBoost', 'Random_Forest', 'SVM', 'MLP']
accuracies = [0.39822587303771606, 0.9080888888888888, 0.8200414950936286, 0.696046662346079, 0.5338]
colors = ['red', 'green', 'blue', 'cyan'] 

plt.figure(figsize=(7, 6))
plt.bar(models, accuracies, color=colors)
plt.title('Accuracy Comparison of Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.ylim(0.0, 1.0)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(models, accuracies, color='blue', marker='*', s=100)
plt.xlabel('Models')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Accuracies')
plt.title('Model Accuracies')

plt.show()
```
